```python
soloviev_data = pd.read_csv('soloviev.csv', index_col=0)
soloviev_data['Date'] = soloviev_data['Date'].apply(lambda x: x.split()[0])
```

```python
soloviev_data = soloviev_data[soloviev_data['Comments'] > 900]
```

```python
for param, c in [('Likes', 'g'), ('Dislikes', 'r'), ('Views', 'b'), ('Comments', 'y')]:
    plt.figure(num=None, figsize=(10, 10), facecolor='w')
    plt.plot(navalny_data['Date'], np.array(navalny_data[param]), color=c, lw=4)
    plt.title(param + ' per video', fontsize=40)
    plt.xticks(rotation=45) 
    plt.xlabel("Dates", fontsize=20)
    plt.ylabel(param, fontsize=20)
    plt.savefig('navalny_' + param + '_rate.png')
```
